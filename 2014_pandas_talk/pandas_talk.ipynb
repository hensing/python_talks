# Warum pandas?


**Library Highlights** ([Quelle](http://pandas.pydata.org/))  
 - A **fast and efficient DataFrame object** for data manipulation **with integrated indexing**;  
 - Tools for **reading and writing data** between in-memory data structures and **different formats**: CSV and text files, Microsoft Excel, SQL databases, and the fast HDF5 format;  
 - Intelligent **data alignment** and integrated **handling of missing data**: gain automatic label-based alignment in computations and easily manipulate messy data into an orderly form;  
 - Flexible reshaping and pivoting of data sets;  
 - Intelligent **label-based slicing**, **fancy indexing**, and subsetting of large data sets;  
 - Columns can be inserted and deleted from data structures for size mutability;  
 - Aggregating or transforming data with a **powerful group by engine** allowing split-apply-combine operations on data sets;  
 - High performance merging and joining of data sets;  
 - **Hierarchical axis indexing** provides an intuitive way of **working with high-dimensional data in a lower-dimensional data structure**;  
 - **Time series-functionality**: date range generation and frequency conversion, moving window statistics, moving window linear regressions, date shifting and lagging. Even create domain-specific time offsets and join time series without losing data;  
 - Highly optimized for performance, with critical code paths written in Cython or C.  
 - Python with pandas is in use in a wide variety of academic and commercial domains, including Finance, Neuroscience, Economics, Statistics, Advertising, Web Analytics, and more.

## Beispiel: Analyse der CSV der Temparatur verschiedener Computer

```python
import pandas as pd
```

```python
# Beispieldatensatz:
fname = 'temps.log'
!tail -n 5 $fname
```

```python
# Daten aus CSV in DataFrame einlesen:
data = pd.read_csv(fname, sep=',',
                   names=['timestamp', 'hostname', 'temp'], 
                   #index_col=('timestamp', 'hostname'),
                   dtype={'timestamp': int, 'hostname': str, 'temp': int},
                   #parse_dates=['timestamp'],  # doesn't work with timestamps!!!
                   error_bad_lines=True)
type(data)
```

```python
# Unter der Haube: numpy:
type(data.values)
```

## Timestamp (int) -> DatetimeIndex

```python
# timestamp -> datetime:
data['timestamp'].astype('M8[s]').tail(3)
```

```python
data['dtime'] = data['timestamp'].astype('M8[s]')
data.tail(3)
```

```python
data.drop('timestamp', axis=1, inplace=True)
data.tail(3)
```

```python
# reindex mit DatetimeIndex:
data.set_index('dtime', inplace=True)
data.tail(3)
```

# Multi-Indizes / Hierachical indexing

```python
# hostname -> machine type via regex
remap = {r'\AAst\d+': 'aeste',
         r'\Axena\d+': 'xenen',
         r'\Azweig\d+': 'zweige',
         r'\A(?!(Ast|zweig|xen|aest))\w+[\d-]*': 'andere'}

data['type'] = data['hostname'].replace(regex=remap)
data.head()
```

```python
# set type and hostname as additional index
data.set_index('type', append=True, inplace=True)
data.set_index('hostname', append=True, inplace=True)
data.tail()
```

```python
type(data.index)
```

## Indizierung bei MultiIndex:

# Analyse mit Multi-Indizes:

```python
# Mittel über hostname
mean_hostname = data.groupby(level='hostname').mean()
mean_hostname.head(3)
```

```python
# Mittel über hosttype
mean_type = data.groupby(level='type').mean()
mean_type.head()
```

```python
# Mittel über hosts pro hosttype, zeitaufgelöst
mean_dtime_type = data.groupby(level=('dtime', 'type')).mean()
mean_dtime_type.tail()
```

# Resampling / rolling mean:

```python
# Mittel über hosts pro hosttype, zeitaufgelöst
# mean_dtime_type = data.groupby(level=('dtime', 'type')).mean()

mean_dtime_type[:1000].unstack()['temp'].plot()
```

```python
# resampling mit Abtastung 1 / 100s
subset = mean_dtime_type[:1000].unstack()['temp']
subset = subset.resample("100s", fill_method="ffill")
subset.plot()
```

```python
# kleinere Fluktuationen uninteressant -> rolling mean:
# Fensterlänge 36 -> 36 * 100s = 60 min

pd.rolling_mean(subset, window=36, min_periods=12).plot()
```

```python
# gleitender Mittelwert über 2h für ca. einen Monat
subset = mean_dtime_type[:50000].unstack()['temp']
subset = subset.resample("100s", fill_method="ffill")
smoothed = pd.rolling_mean(subset, window=72, min_periods=6)
smoothed.plot()
```

# Tricks: Panel / DataFrames aus Dicts

```python
from collections import defaultdict
ddict = lambda: defaultdict(ddict)
import random

temp = ddict()
```

```python
# Fülle dict(dict(...))
for para1 in ['foo', 'bar', 'blub']:
    for para2 in range(20):
        for para3 in range(8, 15):
            # do some calculation / read files ...
            temp[para1][para2][para3] = random.random()
```

```python
# Konvertiere dict(dict(...)) -> panel
panel = pd.Panel(temp)

# Achsen benennen:
panel.items.name = 'type'
panel.major_axis.name = 'para2'
panel.minor_axis.name = 'para1'

panel
```

## Panels -> DataFrame

```python
# panels bestehen aus DataFrames:
panel['bar'].head()
```

```python
# panel -> DataFrame:
panel.to_frame().head()
```

## Save / Load

```python
# save
panel.to_hdf('fname', 'my_key_in_hdf')

# load
store = pd.HDFStore('fname')
store
```
